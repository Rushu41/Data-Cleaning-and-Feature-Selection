```python
import pandas as pd
```

```python
df = pd.DataFrame({"size":["s","m","l","xl","s","s","m","s","xl","l","l","m"]})
df.head()
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["size"]])
```

```python
df["en_size"] = oe.transform(df[["size"]])
df["en_size"] 
```

```python
df
```

```python
#2nd way
or_data = {"s":0,"m":1,"l":2,"xl":3}
```

```python
df["sixe_en_map"]= df["size"].map(or_data)
df["sixe_en_map"]
```
