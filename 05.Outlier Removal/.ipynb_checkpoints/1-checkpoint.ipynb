```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("titanic.csv")
ds.head()
```

```python
ds.info()
```

```python
ds.describe()
```

```python
sns.boxplot(x ="Fare", data = ds)
plt.show()
```

```python
sns.distplot(ds["Fare"])
plt.show()
```

<h2>Outlier Removal</h2>

```python
q1 = ds["Fare"].quantile(0.25)
q3 = ds["Fare"].quantile(0.75)
```

```python
IQR = q3 -q1
```

```python
min_range = q1 - 1.5*IQR
max_range = q3 + 1.5*IQR
max_range,min_range
```

```python
new_ds = ds[ds["Fare"] <= max_range]
```

```python
sns.boxplot(x ="Fare", data = new_ds)
plt.show()
```

<h2>Outlier Removal Using Z Score</h2>

```python
# Z range is -3 to 3
ds1 = pd.read_csv("titanic.csv")
ds1.head()
```

```python
sns.boxplot(x ="Fare", data = ds1)
plt.show()
```

```python
min_range1 = ds["Fare"].mean() - 3*ds["Fare"].std()
max_range1 = ds["Fare"].mean() + 3*ds["Fare"].std()
min_range1,max_range1
```

```python
new_ds1 = ds[ds["Fare"] <= max_range1]
```

```python
sns.boxplot(x ="Fare", data = new_ds1)
plt.show()
```

```python
z_score =( ds1["Fare"] - ds1["Fare"].mean())/ds1["Fare"].std()
z_score
```

```python
ds1["z_score"] = z_score
```

```python
ds[ds1["z_score"] < 3] 
```
