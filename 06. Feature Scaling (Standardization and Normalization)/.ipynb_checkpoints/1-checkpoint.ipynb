```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Standardization

```python
ds = pd.read_csv("titanic.csv")
ds.head()
```

```python
ds.isnull().sum()
```

```python
sns.distplot(ds["Fare"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(ds[["Fare"]])
```

```python
ds["Fare_en"] = pd.DataFrame(ss.transform(ds[["Fare"]]),columns=["x"])
```

```python
ds.head()
```

```python
ds.describe()
```

```python
sns.distplot(ds["Fare_en"])
plt.show()
```
